```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
y = df['diagnosis']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(y)
```

```python
Y
```

```python
df['diagnosis'] = Y
df
```

```python
X = df.drop(['id',  'Unnamed: 32'], axis=1)
X
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
X.shape
```

```python
neurons_n = 10
neurons_m = 10
```

```python
from minisom import MiniSom
```

```python
som = MiniSom(neurons_n, neurons_m, X.shape[1], sigma=0.3, learning_rate=0.5, neighborhood_function='gaussian', random_seed=0)
```

```python
som.pca_weights_init(X)
som.train(X, 1000, verbose=True)  # random training
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(15, 15))
frequencies = som.activation_response(X)
plt.pcolor(frequencies.T, cmap='Greens') 
plt.colorbar()
plt.show()
```
