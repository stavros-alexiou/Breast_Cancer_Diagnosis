```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Repo/data.csv')
```

```python
df
```

```python
y = df['diagnosis'].values
```

```python
y
```

```python
# encoding categorical data from text (B, M) to integers (0, 1)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(y)
```

```python
Y = Y.reshape(-1, 1)
```

```python
Y.ndim
```

```python
# define and normalize/scale values
X = df.drop(['id', 'diagnosis', 'Unnamed: 32'], axis=1)
```

```python
len(X)
```

```python
# scale/normalize the values to bring them to similar range
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
scaler.fit(Y)
Y = scaler.transform(Y)
```

```python
X.ndim
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=42)
```

```python
# check the shape of the traing and testing data
print('Shape of training data is: ', X_train.shape) 
print('Shape of testing data is: ', X_test.shape)
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, CuDNNLSTM, LSTM, Embedding, SimpleRNN
```

```python
model = Sequential()

model.add(Embedding(input_dim=2, output_dim=2, mask_zero=True, input_length=30, trainable=False, embeddings_initializer=tf.keras.initializers.random_normal()))

model.add(LSTM(128, activation='relu'))

model.add(Dropout(0.2))

model.add(Dense(1, activation='sigmoid')) 

model.summary()
```

```python
from keras.backend import binary_crossentropy
model.compile(loss=binary_crossentropy, optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5), metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, epochs=5, validation_data=(X_test, Y_test), validation_split=0.1, verbose=1)
```
