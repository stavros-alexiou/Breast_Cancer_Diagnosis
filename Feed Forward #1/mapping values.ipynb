```python
# create a new csv (coded_values.csv) replacing the values with their code value
```

```python
import pandas as pd
```

```python
df = pd.read_csv('breast-cancer.csv', encoding='utf-8')
```

```python
df
```

```python
map = pd.read_csv('mapping.csv', encoding='utf-8')
```

```python
map
```

```python
values = map.value.to_list()
```

```python
values
```

```python
codes = map.coding.to_list()
```

```python
codes
```

```python
df1 = df.replace(to_replace=values, value=codes)
```

```python
df1
```

```python
df1 = df1.replace(to_replace=['?'], value=[''])
```

```python
df1 = df1.replace(to_replace=[1, 2, 3, '?'], value=['100', '101', '110', ''])
```

```python
df1
```

```python
df1.to_csv('encoded.csv', index=False)
```
